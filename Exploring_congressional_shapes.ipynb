```python
state_fips = pd.read_csv('state_fips.csv')
state_fips.head()
```

```python
sample = pd.read_csv('~/.interchange/gerrymander_sample.csv')
sample = sample.loc[:, ~sample.columns.str.contains('^Unnamed')]

sample.describe()
```

```python
sample = sample.merge(state_fips, left_on='state', right_on='stname')
sample = sample.rename(columns=lambda x: x.strip())
sample.head()
```

```python
r_edge_by_district = sample.groupby(['district', 'st']).r_edge.sum().reset_index()
r_edge_by_district.describe()
```

```python
sample.columns
```

```python
r_edge_by_district['DISTRICTID'] = r_edge_by_district.district.apply(lambda x: f'{x:02}')
r_edge_by_district['STATEID'] = r_edge_by_district['st'].apply(lambda x: f'{x:02}')
r_edge_by_district['GEOID'] =  r_edge_by_district['STATEID'] + r_edge_by_district['DISTRICTID']
sorted(r_edge_by_district.GEOID.unique())
```
